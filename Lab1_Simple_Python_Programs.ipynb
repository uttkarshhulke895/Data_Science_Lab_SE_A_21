### 1. Program to compute the factorial of a given number

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

# Example usage:
number = 5
print(f"The factorial of {number} is {factorial(number)}")

number = 0
print(f"The factorial of {number} is {factorial(number)}")

number = 7
print(f"The factorial of {number} is {factorial(number)}")
```

### 2. Program to display the multiplication table for a given number

```python
def display_table(n):
    print(f"Multiplication Table for {n}:")
    for i in range(1, 11):
        print(f"{n} x {i} = {n*i}")

# Example usage:
number = 7
display_table(number)

print("\n")
number = 12
display_table(number)
```

### 3. Program to identify if a given number is odd or even

```python
def check_odd_even(n):
    if n % 2 == 0:
        print(f"{n} is an even number.")
    else:
        print(f"{n} is an odd number.")

# Example usage:
number = 4
check_odd_even(number)

number = 7
check_odd_even(number)

number = 0
check_odd_even(number)
```

### 4. Program to identify if the entered number is a palindrome or not

```python
def is_palindrome(n):
    # Convert number to string to easily reverse it
    s = str(n)
    if s == s[::-1]:
        print(f"{n} is a palindrome.")
    else:
        print(f"{n} is not a palindrome.")

# Example usage:
number = 121
is_palindrome(number)

number = 123
is_palindrome(number)

number = 1221
is_palindrome(number)

number = 9
is_palindrome(number)
```

### 5. Develop an Expert System that provides simple decision-making

Here's a simple example of an expert system that suggests a beverage based on the user's mood and weather.

```python
def beverage_expert_system():
    print("Welcome to the Beverage Expert System!")
    mood = input("How are you feeling today? (e.g., happy, sad, tired, relaxed): ").lower()
    weather = input("What's the weather like? (e.g., sunny, rainy, cold, hot): ").lower()

    if 'hot' in weather and ('thirsty' in mood or 'tired' in mood):
        print("I recommend a refreshing Iced Tea or Lemonade.")
    elif 'cold' in weather and ('tired' in mood or 'sad' in mood):
        print("A warm Coffee or Hot Chocolate would be great.")
    elif 'sunny' in weather and 'happy' in mood:
        print("Perhaps a nice Smoothie or a sparkling water.")
    elif 'rainy' in weather and 'relaxed' in mood:
        print("A calming Herbal Tea sounds perfect.")
    else:
        print("I'm not sure what to recommend based on your input, but enjoy whatever you choose!")

# Run the expert system
beverage_expert_system()
```
